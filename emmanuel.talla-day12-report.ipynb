### Exercise 1:
Create a simple line plot for `y = x` for `x` values ranging from 0 to 10.

````python
# Create x values from 0 to 10
x = np.linspace(0, 10, 100)

# Compute the four functions
y1 = x
y2 = x**2
y3 = x**3
y4 = x**4

# Create a 2x2 subplot grid
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Plot y1 = x in the first subplot
axs[0, 0].plot(x, y1, color='blue')
axs[0, 0].set_title('y = x')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')

# Plot y2 = x^2 in the second subplot
axs[0, 1].plot(x, y2, color='green')
axs[0, 1].set_title('y = x^2')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')

# Plot y3 = x^3 in the third subplot
axs[1, 0].plot(x, y3, color='orange')
axs[1, 0].set_title('y = x^3')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')

# Plot y4 = x^4 in the fourth subplot
axs[1, 1].plot(x, y4, color='red')
axs[1, 1].set_title('y = x^4')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
````

### Exercise 21:
Plot a contour plot for the function `z = sin(x) + cos(y)` over a grid of `x` and `y` values ranging from -5 to 5.


````python
# Create a grid of x and y values
x = np.linspace(-5, 5, 100)  # 100 points from -5 to 5
y = np.linspace(-5, 5, 100)  # 100 points from -5 to 5
X, Y = np.meshgrid(x, y)  # Create a meshgrid

# Compute z values based on the function z = sin(x) + cos(y)
Z = np.sin(X) + np.cos(Y)

# Create the contour plot
plt.figure(figsize=(10, 6))
contour = plt.contour(X, Y, Z, levels=20, cmap='viridis')  # 20 contour lines with 'viridis' colormap
plt.colorbar(contour)  # Add a colorbar to indicate the value of z

# Add title and labels
plt.title('Contour Plot of z = sin(x) + cos(y)')
plt.xlabel('X')
plt.ylabel('Y')

# Display the plot
plt.show()
````

### Exercise 26:
Create a 3D plot of the function `z = x^2 + y^2` over a grid of `x` and `y` values.


````python
# Generate the x and y values over a grid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Compute the corresponding z values
z = x**2 + y**2

# Create a figure and a 3D axis
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D surface
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Add titles and labels
ax.set_title('3D Plot of z = x^2 + y^2')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Show the plot
plt.show()
````

### Exercise 27:
Create a radar chart with 5 variables and random values between 0 and 1.


````python
# Number of variables (5)
num_vars = 5

# Generate random values between 0 and 1 for each variable
values = np.random.rand(num_vars)

# Create a list of labels for each variable
labels = ['Variable 1', 'Variable 2', 'Variable 3', 'Variable 4', 'Variable 5']

# Add the first value to the end to close the radar chart
values = np.append(values, values[0])

# Calculate the angle for each axis (equal spacing for 5 variables)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Close the circle by appending the first angle at the end

# Create the radar chart
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)

# Draw one axis per variable and one line for each value
ax.fill(angles, values, color='blue', alpha=0.25)
ax.plot(angles, values, color='blue', linewidth=2)

# Add labels to the axes
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

# Set the range of values for the radar chart
ax.set_ylim(0, 1)

# Add a title
plt.title('Radar Chart with 5 Variables', size=15, color='blue')

# Show the plot
plt.show()
````

### Exercise 42:
Create a wind rose chart using random wind direction and speed data.


````python
# Generate random wind direction (0-360 degrees) and speed (0-10 m/s)
num_data_points = 1000
wind_directions = np.random.uniform(0, 360, num_data_points)
wind_speeds = np.random.uniform(0, 10, num_data_points)

# Create the wind rose chart
plt.figure(figsize=(10, 10))

# Create bins for wind direction and speed
bins = np.arange(0, 361, 30)  # 30-degree bins
speed_bins = np.arange(0, 11, 1)  # 1 m/s bins

# Create the histogram in polar coordinates
ax = plt.subplot(111, polar=True)
hist, _, _ = np.histogram2d(wind_directions, wind_speeds, bins=[bins, speed_bins])

# Convert to a radial format
theta = np.deg2rad(bins[:-1])  # Convert bins to radians
radii = hist.sum(axis=1)  # Sum over wind speed bins

# Plot the wind rose
bars = ax.bar(theta, radii, width=np.deg2rad(30), color='green', alpha=0.7)

# Add title and labels
plt.title('Wind Rose Chart')
ax.set_theta_zero_location('N')  # Set North at the top
ax.set_theta_direction(-1)  # Counter-clockwise direction

# Display the plot
plt.show()
````

### Exercise 43:
Plot a cumulative frequency curve for a dataset of 100 random numbers.


````python
# Generate 100 random numbers
data = np.random.randn(100)

# Create a cumulative frequency distribution
counts, bin_edges = np.histogram(data, bins=10, density=True)

# Calculate the cumulative frequency
cumulative = np.cumsum(counts * np.diff(bin_edges))

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(bin_edges[1:], cumulative, marker='o', linestyle='-', color='blue')

# Add titles and labels
plt.title('Cumulative Frequency Curve')
plt.xlabel('Value')
plt.ylabel('Cumulative Frequency')

# Display grid
plt.grid()

# Show the plot
plt.show()
````

### Exercise 44:
Create a waterfall chart using Matplotlib to show the changes in a sequence of values.


````python
# Sample data
categories = ['Start', 'Increase', 'Decrease', 'Increase', 'End']
values = [100, 30, -20, 50, 0]  # Cumulative changes
cumulative = np.zeros(len(values))

# Calculate the cumulative values for the waterfall chart
for i in range(len(values)):
    if i == 0:
        cumulative[i] = values[i]
    else:
        cumulative[i] = cumulative[i - 1] + values[i]

# Create the waterfall chart
fig, ax = plt.subplots(figsize=(8, 6))

# Set colors based on increase/decrease
colors = ['green' if v >= 0 else 'red' for v in values]

# Plot the bars
ax.bar(categories, cumulative, color=colors, alpha=0.7)

# Add the value labels above the bars
for i in range(len(cumulative)):
    ax.text(i, cumulative[i], f'{cumulative[i]:.0f}', ha='center', va='bottom' if cumulative[i] >= 0 else 'top')

# Add lines to show the changes
for i in range(1, len(cumulative)):
    ax.plot([i - 1, i], [cumulative[i - 1], cumulative[i]], color='black', linestyle='--')

# Add title and labels
ax.set_title('Waterfall Chart Example')
ax.set_ylabel('Value')
ax.set_xlabel('Categories')

# Display the plot
plt.grid(axis='y')
plt.show()
````
